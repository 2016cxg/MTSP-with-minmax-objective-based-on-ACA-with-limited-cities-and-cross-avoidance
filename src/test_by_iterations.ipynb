```python

# adjust_edge store edge of two terminal points
# adjust_edge_length store length between two terminal points thought middle points



import numpy as np
import math

datasets = "/home/cheng/PycharmProjects/TSP/datasets/eil51"

data = np.loadtxt( datasets )[ : , 1:]
n = data.shape[0]

D = np.zeros( ( n, n ))
for i in range( 0, n  ):
    for j in range( 0, n ):
        if i != j :
            D[i, j ] = math.sqrt( math.fsum( ( data[i] - data[j]) ** 2 ))
        else :
            D[i, j ] = 0
print( D )

adjust_edge = [ -1 for i in range( n  + 5 ) ]
adjust_edge_length = [ -1 for i in range( n + 5 ) ]
print( adjust_edge_length )
print( adjust_edge )

for i in range( len( all_class ) ):
    adjust_edge[ int( all_class[i][0] ) ] = int( all_class[i][-1] )
    adjust_edge[ int( all_class[i][-1] ) ] = int( all_class[i][0] )
    if len( all_class[i] ) == 1:
        adjust_edge_length[ all_class[i][0] ] = 0 
    else :
        lst = all_class[i]
        tot = 0 
        length = len( all_class[i] )
        for i in range( length -1 ):
            tot = tot + D[ int( lst[i] ) , int( lst[i + 1 ] ) ]
        adjust_edge_length[ int(lst[0] ) ] = tot
        adjust_edge_length[ int(lst[ length - 1] ) ] = tot
print( adjust_edge_length )
print( adjust_edge )
# def dfs( )
```

```python
# for every position i untill all_groups
#   try every city named j, if j has been used in previous positions, skip it
#   till all_groups, calculate the length and compare it with the best solution, 
#       if less than the best solution, replace best solution with current solution
#       otherwise, skip
all_cities = []
all_class_size = len( all_class )

best_solution = [ -1 for i in range( all_class_size )]
best_solution_length = 1e8
tmp_solution = [ -1 for i in range( all_class_size )]
tmp_solution_length = 0

for i in range(all_class_size ):
    if len( all_class[i] ) == 1:
        all_cities.append( all_class[i][0] )
    else :
        all_cities.append( all_class[i][0] )
        all_cities.append( all_class[i][-1] )
print( all_cities )

all_cities_size = len( all_cities )
all_cities_bool = [ 0 for i in range( n + 5)]
def creat_solution( i ):
    global best_solution_length
    global best_solution
    
    # print( i )
    if i >= all_class_size:
        # print("----------------------")
        # print( " i in if %d " % i )
        # print( " tmp solution ")
        # print( tmp_solution )
        length = 0
        for j in range( all_class_size - 1):
            # length pluses length of route part, length between route part j and j + 1
            length = length + adjust_edge_length[ int( tmp_solution[j])]
            length = length + D[ int( adjust_edge[
                                 int( tmp_solution[j] ) ]) ][ int( tmp_solution[j + 1] ) ]
            
            # print( adjust_edge_length[ int( tmp_solution[j] )])
            # print( adjust_edge[ int( tmp_solution[j] ) ] )
            
            
        # length pluses length of last route part and length between last one and first one
        length = length + adjust_edge_length[ int( tmp_solution[ -1 ] ) ]
        length = length + D[ int( adjust_edge[
                             int( tmp_solution[ -1 ] )])][ int( tmp_solution[ 0 ] ) ]
        # print( adjust_edge_length[ int( tmp_solution[ -1 ] )])
        # print( adjust_edge[ int( tmp_solution[ -1 ] ) ] )
        
        if length < best_solution_length:
            best_solution_length = length
            best_solution = tmp_solution
            print("length ")
            print( length )
            print(" best solution length" )
            print( best_solution_length )
            print( best_solution )       
        # print("==========================")
        return 
    
    for j in range( all_cities_size ):
        if all_cities_bool[ int( all_cities[j] )] == 0 and \
            all_cities_bool[ int( adjust_edge[ int( all_cities[j] ) ] ) ] == 0:
            
            # determine value in position i in tmp solution
            tmp_solution[i] = all_cities[j]
            all_cities_bool[ int( all_cities[j] )] = 1
            
            # next position
            creat_solution( i + 1)
            
            # recursive
            all_cities_bool[ int( all_cities[j] )] = 0

creat_solution( 0 )
print("done")
```
